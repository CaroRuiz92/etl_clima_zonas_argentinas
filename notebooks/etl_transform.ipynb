Estadisticas_normales_Datos_abiertos_1991-2020

```python
estad_norm_1991_2020 = pd.read_excel(r'C:\Users\icrui\Documents\DS-ML\proy_etl_clim_arg\data\Estadísticas_normales_Datos_abiertos _1991-2020.xlsx', skiprows=4)
```

```python
estad_norm_1991_2020.head()
```

La forma de la tabla podría no ser provechosa. Se procede a editarla en forma vertical.

```python
estad_norm_1991_2020['Estación'].unique()
```

Se eligen estaciones representativas en ambos datasets, para hacer futuros análisis enfocalizados en ellas.

```python
data_est_eleg_1991_2020 = estad_norm_1991_2020[estad_norm_1991_2020['Estación'].isin(['SALTA AERO', 'IGUAZÚ AERO', 'MENDOZA AERO', 'ROSARIO AERO', 
                                                                                      'AEROPARQUE AERO', 
                                                                                      'BARILOCHE AERO', 'BASE ESPERANZA'])]
```

```python
estac_eleg_smn = estaciones_smn[estaciones_smn['NOMBRE'].isin(['SALTA AERO', 'IGUAZU AERO', 'MENDOZA AERO', 'ROSARIO AERO', 
                                                                                      'AEROPARQUE AERO', 
                                                                                      'BARILOCHE AERO', 'BASE ESPERANZA'])]
```

Se arman datasets más pequeños para su normalización en formato vertical.

```python
raw_data_salta = data_est_eleg_1991_2020[data_est_eleg_1991_2020['Estación'] == 'SALTA AERO']
raw_data_iguazu = data_est_eleg_1991_2020[data_est_eleg_1991_2020['Estación'] == 'IGUAZÚ AERO']
raw_data_mza = data_est_eleg_1991_2020[data_est_eleg_1991_2020['Estación'] == 'MENDOZA AERO']
raw_data_ros = data_est_eleg_1991_2020[data_est_eleg_1991_2020['Estación'] == 'ROSARIO AERO']
raw_data_aeroparq = data_est_eleg_1991_2020[data_est_eleg_1991_2020['Estación'] == 'AEROPARQUE AERO']
raw_data_barilo = data_est_eleg_1991_2020[data_est_eleg_1991_2020['Estación'] == 'BARILOCHE AERO']
raw_data_base_esp = data_est_eleg_1991_2020[data_est_eleg_1991_2020['Estación'] == 'BASE ESPERANZA']
```

```python
# Columnas meses -- se van a usar en todos los sub-datasets
meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']

# Se transforman los meses en filas
df_pivot_salta = raw_data_salta.melt(
    id_vars=['Estación', 'Valor Medio de'],
    value_vars=meses,
    var_name='Mes',
    value_name='Valor')

# De "Valor Medio de" a las columnas
df_final_salta = df_pivot_salta.pivot_table(
    index=['Estación', 'Mes'],
    columns='Valor Medio de',
    values='Valor'
).reset_index()

df_final_salta.columns.name = None # Se elimina el nombre de la columna del índice
```

```python
df_pivot_iguazu = raw_data_iguazu.melt(
    id_vars=['Estación', 'Valor Medio de'],
    value_vars=meses,
    var_name='Mes',
    value_name='Valor')

df_final_iguazu = df_pivot_iguazu.pivot_table(
    index=['Estación', 'Mes'],
    columns='Valor Medio de',
    values='Valor'
).reset_index()

df_final_iguazu.columns.name = None
```

```python
raw_data_mza.replace('S/D', np.nan, inplace=True)
```

```python
df_pivot_mza = raw_data_mza.melt(
    id_vars=['Estación', 'Valor Medio de'],
    value_vars=meses,
    var_name='Mes',
    value_name='Valor')

df_final_mza = df_pivot_mza.pivot_table(
    index=['Estación', 'Mes'],
    columns='Valor Medio de',
    values='Valor'
).reset_index()

df_final_mza.columns.name = None
```

```python
df_pivot_ros = raw_data_ros.melt(
    id_vars=['Estación', 'Valor Medio de'],
    value_vars=meses,
    var_name='Mes',
    value_name='Valor')

df_final_ros = df_pivot_ros.pivot_table(
    index=['Estación', 'Mes'],
    columns='Valor Medio de',
    values='Valor'
).reset_index()

df_final_ros.columns.name = None
```

```python
df_pivot_ap = raw_data_aeroparq.melt(
    id_vars=['Estación', 'Valor Medio de'],
    value_vars=meses,
    var_name='Mes',
    value_name='Valor')

df_final_aeropar = df_pivot_ap.pivot_table(
    index=['Estación', 'Mes'],
    columns='Valor Medio de',
    values='Valor'
).reset_index()

df_final_aeropar.columns.name = None
```

```python
df_pivot_barilo = raw_data_barilo.melt(
    id_vars=['Estación', 'Valor Medio de'],
    value_vars=meses,
    var_name='Mes',
    value_name='Valor')

df_final_barilo = df_pivot_barilo.pivot_table(
    index=['Estación', 'Mes'],
    columns='Valor Medio de',
    values='Valor'
).reset_index()

df_final_barilo.columns.name = None
```

```python
raw_data_base_esp.replace('S/D', np.nan, inplace=True)
```

```python
df_pivot_base_esp = raw_data_base_esp.melt(
    id_vars=['Estación', 'Valor Medio de'],
    value_vars=meses,
    var_name='Mes',
    value_name='Valor')

df_final_base_esp = df_pivot_base_esp.pivot_table(
    index=['Estación', 'Mes'],
    columns='Valor Medio de',
    values='Valor'
).reset_index()

df_final_base_esp.columns.name = None
```

Se hace concat con los "subdatasets" armados.

```python
data_climas_selec = pd.concat([df_final_salta, df_final_iguazu, df_final_mza, df_final_ros, df_final_aeropar, df_final_barilo, df_final_base_esp], ignore_index=True)
```

```python
data_climas_selec
```
